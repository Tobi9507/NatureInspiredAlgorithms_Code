```python
# imports
# add folder to syspath to import PSO
import os
import sys
path = os.path.dirname(os.path.dirname(os.path.abspath('')))
sys.path.insert(0, path+'/ParticleSwarmOptimization')
from ParticleSwarmOptimization import ParticleSwarmOptimization

# add folder to syspath to import GA
#sys.path.insert(0, os.path.dirname(path)+'/GeneticAlgorithm')
 # add folder to syspath to import ACO 
#sys.path.insert(0, os.path.dirname(path)+'/AntColonyOptimization')

import numpy as np
from matplotlib import pyplot as plt
```

```python
# lade die Probleme
# small problem
small_objects  = np.load(os.path.join(path,'Compare/Ressources/betterproblem/small_objects.npy'))
small_container = np.load(os.path.join(path,'Compare/Ressources/betterproblem/small_container.npy'))
small_optimal_solution = np.load(os.path.join(path, 'Compare/Ressources/betterproblem/small_optimal.npy'))
# medium Problem
medium_objects= np.load(os.path.join(path,'Compare/Ressources/betterproblem/medium_objects.npy'))
medium_container = np.load(os.path.join(path,'Compare/Ressources/betterproblem/medium_container.npy'))
medium_optimal_solution = np.load(os.path.join(path, 'Compare/Ressources/betterproblem/medium_optimal.npy'))
# large Problem
large_objects  = np.load(os.path.join(path,'Compare/Ressources/betterproblem/large_objects.npy'))
large_container = np.load(os.path.join(path,'Compare/Ressources/betterproblem/large_container.npy'))
large_optimal_solution = np.load(os.path.join(path, 'Compare/Ressources/betterproblem/large_optimal.npy'))
# medium Problem neu
medium_1_100_objects = np.load(os.path.join(path,'Compare/Ressources/new_medium1-100.npy'))
medium_1_100_objects.shape = (250,2)
medium_1_100_container = np.load(os.path.join(path,'Compare/Ressources/betterproblem/medium_container.npy'))
medium_20_80_objects = np.load(os.path.join(path,'Compare/Ressources/new_medium20-80.npy'))
medium_20_80_objects.shape = (250,2)
medium_20_80_container = np.load(os.path.join(path,'Compare/Ressources/betterproblem/medium_container.npy'))
medium_40_60_objects = np.load(os.path.join(path,'Compare/Ressources/new_medium40-60.npy'))
medium_40_60_objects.shape = (250,2)
medium_40_60_container = np.load(os.path.join(path,'Compare/Ressources/betterproblem/medium_container.npy'))


print('Optimale Lösungen')
print(f'Kleines Problem {small_optimal_solution}')
print(f'Medium Problem {medium_optimal_solution}')
print(f'Großes Problem {large_optimal_solution}')
```

```python
iterations = 150
number_particles = 40
number_average = 1

a0_solution_bins = np.zeros(number_average)
a1_solution_bins = np.zeros(number_average)
a2_solution_bins = np.zeros(number_average)

a0_runtime = np.zeros(number_average)
a1_runtime = np.zeros(number_average)
a2_runtime = np.zeros(number_average)

a0_average_bins = np.zeros((number_average, iterations))
a1_average_bins = np.zeros((number_average, iterations))
a2_average_bins = np.zeros((number_average, iterations))

a0_best_bins = np.zeros((number_average, iterations))
a1_best_bins = np.zeros((number_average, iterations))
a2_best_bins = np.zeros((number_average, iterations))

a0_worst_bins = np.zeros((number_average, iterations))
a1_worst_bins = np.zeros((number_average, iterations))
a2_worst_bins = np.zeros((number_average, iterations))

a0_fitness = np.zeros((number_average, iterations))
a1_fitness = np.zeros((number_average, iterations))
a2_fitness = np.zeros((number_average, iterations))

a0_unfit_moves = np.zeros((number_average, iterations))
a1_unfit_moves = np.zeros((number_average, iterations))
a2_unfit_moves = np.zeros((number_average, iterations))

a0_chaos_moves = np.zeros((number_average, iterations))
a1_chaos_moves = np.zeros((number_average, iterations))
a2_chaos_moves = np.zeros((number_average, iterations))

a0_heuristic_moves = np.zeros((number_average, iterations))
a1_heuristic_moves = np.zeros((number_average, iterations))
a2_heuristic_moves = np.zeros((number_average, iterations))

for i in range(number_average):
    print(str(i+1), end = ' || ')
    PSO = ParticleSwarmOptimization(number_particles, iterations, medium_1_100_objects, medium_1_100_container[0], medium_1_100_container[1], 0.4, 0.2, 0.2, 0.2, 0.4, 0.2, 'random', 'first_fit', 'first_fit')
    _, a0_solution_bins[i], a0_runtime[i], a0_average_bins[i], a0_best_bins[i], a0_worst_bins[i], a0_fitness[i], a0_unfit_moves[i], a0_chaos_moves[i], a0_heuristic_moves[i] = PSO.run()
    print('a0 : ' + str(round(a0_runtime[i], 1)), end = ' | ')
    PSO = ParticleSwarmOptimization(number_particles, iterations, medium_20_80_objects, medium_20_80_container[0], medium_20_80_container[1], 0.4, 0.2, 0.2, 0.2, 0.4, 0.2, 'random', 'first_fit', 'first_fit')
    _, a1_solution_bins[i], a1_runtime[i], a1_average_bins[i], a1_best_bins[i], a1_worst_bins[i], a1_fitness[i], a1_unfit_moves[i], a1_chaos_moves[i], a1_heuristic_moves[i] = PSO.run()
    print('a1 : ' + str(round(a1_runtime[i], 1)), end = ' | ')
    PSO = ParticleSwarmOptimization(number_particles, iterations, medium_40_60_objects, medium_40_60_container[0], medium_40_60_container[1], 0.4, 0.2, 0.2, 0.2, 0.4, 0.2, 'random', 'first_fit', 'first_fit')
    _, a2_solution_bins[i], a2_runtime[i], a2_average_bins[i], a2_best_bins[i], a2_worst_bins[i], a2_fitness[i], a2_unfit_moves[i], a2_chaos_moves[i], a2_heuristic_moves[i] = PSO.run()
    print('a2 : ' + str(round(a2_runtime[i], 1)))
    
# Durschnitt berechnen

average_a0_solution_bins = np.average(a0_solution_bins, 0)
average_a0_runtime = np.average(a0_runtime, 0)
average_a0_bins = np.average(a0_average_bins, 0)
average_a0_best = np.average(a0_best_bins, 0)
average_a0_worst = np.average(a0_worst_bins, 0)
average_a0_fitness = np.average(a0_fitness, 0)
average_a0_unfit_moves = np.average(a0_unfit_moves, 0)
average_a0_chaos_moves = np.average(a0_chaos_moves, 0)
average_a0_heuristic_moves = np.average(a0_heuristic_moves, 0)

average_a1_solution_bins = np.average(a1_solution_bins, 0)
average_a1_runtime = np.average(a1_runtime, 0)
average_a1_bins = np.average(a1_average_bins, 0)
average_a1_best = np.average(a1_best_bins, 0)
average_a1_worst = np.average(a1_worst_bins, 0)
average_a1_fitness = np.average(a1_fitness, 0)
average_a1_unfit_moves = np.average(a1_unfit_moves, 0)
average_a1_chaos_moves = np.average(a1_chaos_moves, 0)
average_a1_heuristic_moves = np.average(a1_heuristic_moves, 0)

average_a2_solution_bins = np.average(a2_solution_bins, 0)
average_a2_runtime = np.average(a2_runtime, 0)
average_a2_bins = np.average(a2_average_bins, 0)
average_a2_best = np.average(a2_best_bins, 0)
average_a2_worst = np.average(a2_worst_bins, 0)
average_a2_fitness = np.average(a2_fitness, 0)
average_a2_unfit_moves = np.average(a2_unfit_moves, 0)
average_a2_chaos_moves = np.average(a2_chaos_moves, 0)
average_a2_heuristic_moves = np.average(a2_heuristic_moves, 0)

print(average_a0_solution_bins, end = ' | ')
print(average_a1_solution_bins, end = ' | ')
print(average_a2_solution_bins)
print(average_a0_runtime, end = ' | ')
print(average_a1_runtime, end = ' | ')
print(average_a2_runtime)

x_vals = np.arange(iterations)
fig, ax = plt.subplots(5, figsize=(16,16))
fig.tight_layout(h_pad=6)
ax[0].set_title(f'Average Number of bins')
ax[0].plot(x_vals, average_a0_bins, color="blue", label = 'a0')
ax[0].plot(x_vals, average_a1_bins, color="green", label = 'a1')
ax[0].plot(x_vals, average_a2_bins, color="red", label = 'a2')
ax[0].set_xlabel('Iteration')
ax[0].set_ylabel("Average number of Bins")
ax[1].set_title(f'Best bins')
ax[1].plot(x_vals, average_a0_best, color="blue", label = 'a0')
ax[1].plot(x_vals, average_a1_best, color="green", label = 'a1')
ax[1].plot(x_vals, average_a2_best, color="red", label = 'a2')
ax[1].set_xlabel('Iteration')
ax[1].set_ylabel("Average number of Bins")
ax[2].set_title(f'Swarm Fitness')
ax[2].plot(x_vals, average_a0_fitness, color="blue", label = 'a0')
ax[2].plot(x_vals, average_a1_fitness, color="green", label = 'a1')
ax[2].plot(x_vals, average_a2_fitness, color="red", label = 'a2')
ax[2].set_xlabel('Iteration')
ax[2].set_ylabel("Average Fitness")
ax[3].set_title(f'Unfit Object Moves')
ax[3].plot(x_vals, average_a0_unfit_moves, color="blue", label = 'a0')
ax[3].plot(x_vals, average_a1_unfit_moves, color="green", label = 'a1')
ax[3].plot(x_vals, average_a2_unfit_moves, color="red", label = 'a2')
ax[3].set_xlabel('Iteration')
ax[3].set_ylabel("Average number of Moved Objects")
ax[4].set_title(f'Chaos Object Moves')
ax[4].plot(x_vals, average_a0_chaos_moves, color="blue", label = 'a0')
ax[4].plot(x_vals, average_a1_chaos_moves, color="green", label = 'a1')
ax[4].plot(x_vals, average_a2_chaos_moves, color="red", label = 'a2')
ax[4].set_xlabel('Iteration')
ax[4].set_ylabel("Average number of Moved Objects")
plt.legend()
plt.show()
```
