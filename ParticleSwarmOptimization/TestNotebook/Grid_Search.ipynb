```python
iterations = 150
number_particles = 40
number_average = 5

solutions = np.empty((3, 3, 3, 3, 3, number_average))
average_solutions = np.empty((3, 3, 3, 3, 3))
runtime = np.empty((3, 3, 3, 3, 3, number_average))
average_runtime = np.empty((3, 3, 3, 3, 3))
bins_average = np.empty((3, 3, 3, 3, 3, number_average, iterations))

local_ce = [0.3, 0.4, 0.5]
local_ce_change = [0.1, 0.2, 0.3]
global_ce = [0.1, 0.2, 0.3]
global_ce_change = [0.3, 0.4, 0.5]
chaos_ce = [0.1, 0.15, 0.2]
count = 1
                
for i in range(3):
    for j in range(3):
        for k in range(3):
            for l in range(3):
                for m in range(3):
                    for n in range(number_average):
                        PSO = ParticleSwarmOptimization(number_particles, iterations, large_objects, large_container[0], large_container[1], local_ce[i], global_ce[k], chaos_ce[m], local_ce_change[j], global_ce_change[l], chaos_ce[m], 'random', 'first_fit', 'first_fit')
                        _, solutions[i][j][k][l][m][n], runtime[i][j][k][l][m][n], bins_average[i][j][k][l][m][n], _, _, _, _, _, _ = PSO.run()
                    average_solutions[i][j][k][l][m] = np.average(solutions[i][j][k][l][m], 0)
                    average_runtime[i][j][k][l][m] = np.average(runtime[i][j][k][l][m], 0)
                    print(str(count) + ' || l: ' + str(local_ce[i]) + ' | l_c: ' + str(local_ce_change[j]) + ' | g: ' + str(global_ce[k]) + ' | g_c: ' + str(global_ce_change[l]) + ' | c: ' + str(chaos_ce[m]) + ' || average_solution: ' + str(average_solutions[i][j][k][l][m]) + ' | average_runtime: ' + str(average_runtime[i][j][k][l][m]))
                    count += 1
```
